```python
import csv 
import pandas as pd
import numpy as np


data_df = pd.read_pickle("/mnt/c/Users/user/Downloads/data/data.pkl")

df = pd.DataFrame(data_df.iloc[:,1][:].tolist())
print(df.head())
print(df.shape)
#print(len(data_df))
#print(data_df.index)
#print(data_df.dtypes)
#print(data_df.shape)
#print(data_df.head())
#print(data_df.tail())
train_x = df[:5977].to_numpy()
test_x = df[5977:].to_numpy()
train_y = data_df['index'][:5977].to_numpy()
test_y = data_df['index'][5977:].to_numpy()
```

```python

print(train_x.dtype, train_y.dtype, test_x.dtype, test_y.dtype)

np_data=data_df.to_numpy()
print(np_data[0][0], len(np_data[0][1]))

import torch 
from torch import nn
from torch import Tensor
from torch.autograd import Variable

class dnn(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(dnn, self).__init__()
        self.linear = nn.Sequential(
            nn.Dropout(0.2),
            nn.Linear(n_feature, n_hidden),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(n_hidden, n_hidden),
            nn.ReLU(),
            nn.Dropout(0.15),
            nn.Linear(n_hidden, n_hidden),
            nn.ReLU(),
            nn.Dropout(0.10),
            nn.Linear(n_hidden, n_hidden),
            nn.ReLU(),
            nn.Dropout(0.05),
            nn.Linear(n_hidden, n_hidden),
            nn.ReLU(),
            nn.Linear(n_hidden, n_output))

    def forward(self, x):
        out = self.linear(x)
        return out



n_feature = len(np_data[0][1])
n_hidden = 1000
n_out = 1
learningRate = 0.01 
epochs = 100
batch_size = 50

model = dnn(n_feature, n_hidden, n_out)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
loss_func = torch.nn.MSELoss()

#train
loss_all = []
for t in range(100):
    loss_sum = 0
    #for i in range(0,len(train_x), batch_size):
    x = Variable(torch.from_numpy(train_x).float())
    y = Variable(torch.from_numpy(train_y).float())
    #x = torch.reshape(torch.Tensor(train_x),(n_feature))
    #print(x.size())
    #y = torch.Tensor(train_y)
    model.train()
    optimizer.zero_grad()
    prediction = model(x)
    loss = loss_func(prediction, y)
    loss_sum += loss.detach().numpy()
    loss.backward()
    optimizer.step()
    loss_all.append(loss)
    print('epoch = {}, loss = {}'.format(t,loss_sum))
    #optimizer.step()

# predict
model.eval()
x = Variable(torch.from_numpy(test_x).float())
pred_test = model(x)
pred_test
```
